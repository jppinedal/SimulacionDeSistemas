```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Simulaci√≥n de sistemas/lotmed_ran1.xlsx',header = None)

```

```python


lista_total = []

for i in df[0]:
  lista_total.append(i)

ascendente = 0
descendente = 0


for i, num in  enumerate(lista_total):
   if i == len(lista_total) - 1:
    break

   elif i == 0:
      ascendente +=1 #  El primer elemento cuenta como racha

   elif lista_total[i-1] < lista_total[i] < lista_total[i+1]: # todos aumentan
    continue

   elif lista_total[i-1] > lista_total[i] > lista_total[i+1]: # todos disminuyen
    continue

   elif lista_total[i-1] < lista_total[i] & lista_total[i] > lista_total[i+1]: # Cuando aumenta cambia a disminuye
    descendente += 1

   elif lista_total[i-1] > lista_total[i] & lista_total[i] < lista_total[i+1]: # Cuando disminuye cambia a aumenta
     ascendente += 1



print(lista_total)
print("Rachas ascendente: ",ascendente)
print("Rachas descendente: ",descendente)
print("Total de rachas: ",ascendente+descendente)
```

```python
seudoaleatorios_x = []
seudoaleatorios_y = []
copia_lista_total = lista_total[:]
#print(len(copia_lista_total))

while copia_lista_total: #miestras la lista no est√© vac√≠a
  seudoaleatorios_x.append(copia_lista_total.pop(0)/9999) #Tomamos el primer elemento de la lista, le hacemos pop, lo dividimos por el n√∫mero mayor (9999) para que nos quede en rango [0,1] y lo agregamos a la lista que ha de contener la coordenada x
  seudoaleatorios_y.append(copia_lista_total.pop(0)/9999) #Tomamos el primer elemento de la lista, le hacemos pop, lo dividimos por el n√∫mero mayor (9999) para que nos quede en rango [0,1] y lo agregamos a la lista que ha de contener la coordenada y



print(seudoaleatorios_x)  #  xj entre 0 y 1
print(seudoaleatorios_y) #xj+1 entre 0 y 1
print(lista_total)
print(len(copia_lista_total))
print(len(seudoaleatorios_y)) #Tama√±o de la lista x
print(len(seudoaleatorios_x)) #Tama√±o de la lista x


```

```python
# Prueba espectral

# Crear el gr√°fico de dispersi√≥n (scatter plot)
plt.scatter(seudoaleatorios_x, seudoaleatorios_y, color='black', label='Puntos')

# Agregar etiquetas y t√≠tulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Prueba Espectral')

# Mostrar leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```
