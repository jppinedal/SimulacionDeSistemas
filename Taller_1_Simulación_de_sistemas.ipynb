# Nueva secci√≥n

```python
# Punto 1
# librer√≠as
import random
import math

def calcular_distancia(p1, p2):
    suma_cuadrados = 0
    for i in range(len(p1)):
        suma_cuadrados += (p1[i] - p2[i]) ** 2
    return math.sqrt(suma_cuadrados)

def distancia_promedio_en_cubo(numero_pares):
    suma_distancias = []

    for _ in range(numero_pares):
        punto1 = [random.random(), random.random(), random.random()] #Coordenadas x,y, z
        punto2 = [random.random(), random.random(), random.random()]
        distancia = calcular_distancia(punto1, punto2)
        suma_distancias.append(distancia)
    return suma_distancias

lista_promedio = []
lista_std = []

for i in range(20):
  numero_pares_aleatorios = 51
  promedio_distancias = distancia_promedio_en_cubo(numero_pares_aleatorios)
  print(f"Distancia promedio entre {numero_pares_aleatorios} pares de puntos: {promedio_distancias}")
  if i <= 19:
    promedio = sum(promedio_distancias)/len(promedio_distancias)
    std = 0
    for i in promedio_distancias:
      std = (((i - promedio)**2)/(len(promedio_distancias)-1))**(1/2)
    lista_promedio.append(promedio)
    lista_std.append(std)

print(f"Lista de promedios de las distancias de los 51 pares en las 20 ejecuciones {lista_promedio}")
print(f"Promedio de distancias en el total de 20 ejecuciones {sum(lista_promedio)/len(lista_promedio)}")
print(f"Lista de las desviaciones estandar de las distancias de los 51 pares en las 20 ejecuciones {lista_std}")
```

```python
# Intervalo de confianza

promedio_gen_std = sum(lista_std)/len(lista_std)
intervalo_confianza_positivo = (sum(lista_promedio)/len(lista_promedio)) + 2.0930 * promedio_gen_std/(20**(1/2))
intervalo_confianza_negativo = (sum(lista_promedio)/len(lista_promedio)) - 2.0930 *promedio_gen_std/(20**(1/2))

print(f"El intervalo de confianza es [ {intervalo_confianza_negativo} , {intervalo_confianza_positivo}]")
```
